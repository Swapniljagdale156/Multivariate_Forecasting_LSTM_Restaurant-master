```
df_new = df[['TicketCode','PartySize','MenuCateogry','MenuItem', 'ItemPrice', 'ItemQty']]
```

```
def Weekday_Weekend(date):
    if date.weekday() >= 0 and date.weekday() <=3:
        return 'Weekday'
    elif date.weekday() == 4:
        return 'Fri'
    else:
        return 'Weekend'
```

```
import random
random.seed(101)

def expansion_randomize_data(start_date, end_date, date, df_new):

    rows_list = []
    while date >= start_date and date <= end_date:
        row_list = []
        count = 0

        lunch_start = datetime(date.year, date.month, date.day, 11,30,0)
        lunch_end = datetime(date.year, date.month, date.day, 14,0,0)
        dinner_start = datetime(date.year, date.month, date.day, 18,30,0)
        dinner_end = datetime(date.year, date.month, date.day, 22,0,0)

        if Weekday_Weekend(date) == 'Weekday':
            count = random.choice(range(1,6))
        elif Weekday_Weekend(date) == 'Fri' and date <=lunch_end:
            count = random.choice(range(1,6))
        elif Weekday_Weekend(date) == 'Fri' and date <=dinner_end:
            count = random.choice(range(5,10))
        if Weekday_Weekend(date) == 'Weekend':
            count = random.choice(range(5,10))

        for i in range(count):
            row_list = []
            if (date >= lunch_start and date <= lunch_end):
                row_list.append(date)
                row_list.append('lunch')
                row_list.append(date.weekday()+1)
                row_list += list(df_new.iloc[random.choice(range(200))])
                print(row_list)
                rows_list.append(row_list)
            elif (date >= dinner_start and date <=dinner_end):
                row_list.append(date)
                row_list.append('dinner')
                row_list.append(date.weekday()+1)
                row_list += list(df_new.iloc[random.choice(range(200))])
                print(row_list)
                rows_list.append(row_list)

        date += timedelta(minutes=5)    
    return rows_list

start_date = datetime.strptime('2019-01-01 00:00:00', '%Y-%m-%d %H:%M:%S')
end_date = datetime.strptime('2019-06-30 23:00:00', '%Y-%m-%d %H:%M:%S')
date = datetime.strptime('2019-01-01 00:00:00', '%Y-%m-%d %H:%M:%S')
rows_list = expansion_randomize_data(start_date, end_date, date, df_new)
```

```
import pandas as pd 
pd.set_option('display.max_columns', 500)
pd.set_option('display.max_rows', 10000)
df_new = pd.DataFrame(rows_list, columns=['dts','shift','weekday','ticket_code','party_size','menu_category','menu_item', 'item_price', 'item_qty'])
df_new.head()
```
